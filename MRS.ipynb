Useful columns

```python
#id, genres,keywords,title,overview,cast,crew
movie_data = movie_data[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie_data.head(1)
```

**Preprocessing**

```python
movie_data.isnull().sum()
```

```python
movie_data.dropna(inplace=True)
```

```python
movie_data.duplicated().sum()
```

```python
#In order to make tags we have to make data into meaningful foramt
#initially format like
movie_data.iloc[0].genres
```

```python
import ast
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L  
```

```python
movie_data['genres']=movie_data['genres'].apply(convert)
movie_data.head()
```

```python
# same function apply on others colums
# same function apply on others colums
movie_data['keywords']=movie_data['keywords'].apply(convert)
movie_data.head()
```

```python
# now for actors we need at least three actors
import ast
def convertact(obj):
  L=[]
  count=0
  for i in ast.literal_eval(obj):
    if count!=3:
      L.append(i['name'])
      count+=1
    else:
      break
  return L  
```

```python
movie_data['cast']=movie_data['cast'].apply(convertact)
movie_data.head()
```

```python
# now get director from crew
def director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']== 'Director':
      L.append(i['name'])
      break
  return L  
```

```python
movie_data['crew']=movie_data['crew'].apply(director)
movie_data.head()
```

```python
# convert overview string into list
movie_data['overview'] = movie_data['overview'].apply(lambda x:x.split())
movie_data.head()
```

```python
# remove spaces between words
movie_data['genres']=movie_data['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie_data.head()
```

```python
#remove spaces for other three columns
movie_data['genres']=movie_data['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movie_data['cast']=movie_data['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movie_data['keywords']=movie_data['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movie_data['crew']=movie_data['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movie_data.head()
```

**Tags**

```python
movie_data['tags']=movie_data['overview']+movie_data['genres']+movie_data['keywords']+movie_data['cast']+movie_data['crew']
movie_data.head()
```

```python
#now for four columns we'll use only one columns called tags, so make new daatsets
new_data = movie_data[['movie_id','title','tags']]
new_data
```

```python
# now convert tags list into string
new_data['tags']=new_data['tags'].apply(lambda x:" ".join(x))
new_data
```

```python
#convert tags string into lower case
new_data['tags']=new_data['tags'].apply(lambda x:x.lower())
new_data
```

**Vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer
countvector = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=countvector.fit_transform(new_data['tags']).toarray()
vectors
```

```python
countvector.fit_transform(new_data['tags']).toarray().shape
```

**Stemming**

```python
import nltk
from nltk.stem.porter import PorterStemmer
portstemmer=PorterStemmer()
```

```python
#function for stemming
def stemm(text):
  y=[]
  for i in text.split():
    y.append(portstemmer.stem(i))
    return " ".join(y)
```

```python
new_data['tags']=new_data['tags'].apply(stemm)
```

```python
countvector.get_feature_names()
```

**Cosine distance between vectors**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# calculate distance of each vectors with all other
similar_movie = cosine_similarity(vectors)
```

```python
#recommendation functions, get index of movie and show the 5  most similar movies in start while different at end
def recommendation(movie):
  movie_index = new_data[new_data['title']==movie].index[0]
  movie_distances = similar_movie[movie_index]
  list_movies = sorted(list(enumerate(movie_distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in list_movies:
    print(new_data.iloc[i[0]].title)
```

```python
recommendation('Batman Begins')
```
